```python
import pandas as pd
import dateutil.parser as parser
from datetime import datetime
```

```python
excel_file = pd.ExcelFile('A11B8A30.xlsx')
print(excel_file.sheet_names)
```

```python
raw_sheet_data = pd.read_excel(excel_file,'Raw')
estimation_data1 = pd.read_excel(excel_file,'estimations MM (1)')
estimation_data2 = pd.read_excel(excel_file,'estimations MM (2)')
```

```python
print(raw_sheet_data)
```

```python
event_date = raw_sheet_data['31 infotech'][0]
e = event_date.split('.')
event_date = e[1]+'.'+e[0]+'.'+e[2]
event_date = parser.parse(event_date)
```

```python
print(event_date)
```

```python
func = lambda x: datetime.datetime(parser.parse(x)) if isinstance(x,str) else x
```

```python
raw_sheet_data['Date'] = raw_sheet_data['Unnamed: 0'].apply(func)
print(raw_sheet_data['Unnamed: 0'])
```

```python
raw_sheet_data['Date']
```

```python
event_location = int(raw_sheet_data.loc[raw_sheet_data['Date']==pd.to_datetime(event_date)].index[0])
```

```python
raw_sheet_data.loc[event_location]
```

```python
event_window_df = pd.DataFrame()
for i in range(event_location-20,event_location+21):
    event_window_df=event_window_df.append(raw_sheet_data.loc[i],['Date','31 infotech'])
```

```python
estimation_period_df = pd.DataFrame()
for i in range(event_location-120,event_location-20):
    estimation_period_df = estimation_period_df.append(raw_sheet_data.loc[i],['Date','31 infotech'])
print(len(estimation_period_df))
```

```python
event_window_df
```

```python
bse_average = event_window_df['Sensex'].mean()
company_average = event_window_df['31 infotech'].mean()
bse_std = event_window_df['Sensex'].std()
company_std = event_window_df['31 infotech'].std()
print(bse_average,bse_std,company_average,company_std)
```

```python
event_window_df['bse_diff'] = event_window_df['Sensex'] - bse_average
event_window_df['company_diff'] = event_window_df['31 infotech'] - company_average
```

```python
event_window_df['diff_squared'] = event_window_df['bse_diff'] ** 2
event_window_df['diff_product'] = event_window_df['bse_diff']*event_window_df['company_diff']
beta = event_window_df['diff_product'].sum()/event_window_df['diff_squared'].sum()
alpha = company_average - (beta*bse_average)
event_window_df['expected_return'] = alpha + (beta * event_window_df['Sensex'])
event_window_df['abnormal_return'] = event_window_df['31 infotech'] - event_window_df['expected_return']
print(beta)
print(alpha)
```

```python
estimation_period_df
```

```python
bse_average_est = estimation_period_df['Sensex'].mean()
bse_std_est = estimation_period_df['Sensex'].std()
company_avg_est = estimation_period_df['31 infotech'].mean()
company_std_est = estimation_period_df['31 infotech'].std()
```

```python
print(bse_std_est,company_std_est)
```

```python
estimation_period_df['bse_diff'] = estimation_period_df['Sensex'] - bse_average_est
estimation_period_df['cmp_diff'] = estimation_period_df['31 infotech'] - company_avg_est
estimation_period_df['diff_squared'] = estimation_period_df['bse_diff'] ** 2
estimation_period_df['diff_product'] = estimation_period_df['bse_diff'] * estimation_period_df['cmp_diff']
beta = estimation_period_df['diff_product'].sum()/estimation_period_df['diff_squared'].sum()
alpha = company_avg_est - (beta * bse_average_est)
estimation_period_df['expected_return'] = alpha + (beta * estimation_period_df['Sensex'])
estimation_period_df['abnormal_return'] = estimation_period_df['31 infotech'] - estimation_period_df['expected_return']
```

```python
print(alpha, beta)

print(estimation_period_df['abnormal_return'].std())
```

```python
excel_name = 'Results.xlsx'
xl = pd.ExcelWriter(excel_name,engine='xlsxwriter')
event_window_df.to_excel(xl,sheet_name='Event Window')
estimation_period_df.to_excel(xl,sheet_name='Estimation Period')
xl.save()
xl.close()
```
